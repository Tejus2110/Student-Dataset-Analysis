### Q10 - How many students from various cities? (solve by visualisation)

```python
group_four = df.groupby('city')
students_from_diff_city = group_four['first name'].agg('count')

students_var = students_from_diff_city.reset_index()
print(students_var.sort_values(by = ['first name'] ,ascending = False)[0:5]) # Top 5 cities
students_var.columns = ['City', 'Count']

# sorting for better visualisation
students_var = students_var.sort_values(by = ['Count'], ascending = False)
# Create a treemap
plt.figure(figsize=(20, 12))
squarify.plot(sizes=students_var['Count'], label=students_var['City'], alpha=0.8)
plt.axis('off')
plt.title('Number of Students from Various Cities')
plt.show()
```

### Q11 - How does the expected salary vary based on factors like GPA and Family Income and Experience with python in months?

```python
df_analysis = df[['cgpa', 'family_income_calculated', 'experience with python (months)', 'expected salary (lac)']]


df_analysis['cgpa'] = pd.to_numeric(df_analysis['cgpa'], errors='coerce')
df_analysis['family_income_calculated'] = pd.to_numeric(df_analysis['family_income_calculated'], errors='coerce')
df_analysis['experience with python (months)'] = pd.to_numeric(df_analysis['experience with python (months)'], errors='coerce')
df_analysis['expected salary (lac)'] = pd.to_numeric(df_analysis['expected salary (lac)'], errors='coerce')



# Set up the plotting style


# Plot 1: Expected Salary vs CGPA
plt.figure(figsize=(10, 6))
sns.scatterplot(x='cgpa', y='expected salary (lac)', data=df_analysis)
plt.title('Expected Salary vs CGPA')
plt.xlabel('CGPA')
plt.ylabel('Expected Salary (in Lacs)')
plt.show()

# Plot 2: Expected Salary vs Family Income
plt.figure(figsize=(10, 6))
sns.scatterplot(x='family_income_calculated', y='expected salary (lac)', data=df_analysis)
plt.title('Expected Salary vs Family Income')
plt.xlabel('Family Income Calculated')
plt.ylabel('Expected Salary (in Lacs)')
plt.show()

# Plot 3: Expected Salary vs Experience with Python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='experience with python (months)', y='expected salary (lac)', data=df_analysis)
plt.title('Expected Salary vs Experience with Python (Months)')
plt.xlabel('Experience with Python (Months)')
plt.ylabel('Expected Salary (in Lacs)')
plt.show()

# Optional: Pair Plot to visualize the relationships among key variables
sns.pairplot(df_analysis)
plt.suptitle('Pair Plot of Key Variables', y=1.02)
plt.show()
```

Solution:
- The analysis suggests that among the three factors, CGPA has a relatively stronger association with expected salary, but even this is not definitive due to the variability within each CGPA range.
- Family income and Python experience (in months) show weak or no clear relationships with expected salary.
- Overall, while CGPA might play some role in influencing expected salary, other unmeasured factors could be playing significant roles,  
such as skill sets, certifications, internships, or soft skills.

### Q13 - Do students in leadership positions during their college year tend to have higher GPAs or better expected salary?

Solution:- 
- The median cgpa of students with leadership and without leadership skills is same. 
- The median expected salary of the students with leadership skills is higher as compared to the students without leadership skills.

Conclusion: 
- The students with leadership skill tend to have higher expected salary than students without leadership skills.
- But it doesn't affect the cgpa

### Q16 - Find total number of students who attended the events related to Data Science

```python
print(df['events'].value_counts())
# from this we came to know these course are related to data science are as follows Data Visualization using Power BI ,Hello ML and DL and Artificial Intelligence
# create a function which will check the value of the cell and if its is one of the three then we will count
counter = 0
checker_list = ['Data Visualization using Power BI','Hello ML and DL','Artificial Intelligence']
def counter_func(cell):
    global counter
    if cell in checker_list:
        counter += 1
# passing the function to the events column
df['events'].map(counter_func)
print('The number of students who attended courses related to Data Science ' , counter)
```

Solution:

Total students who attended courses related to Data Science are 582.

### Q17 - Those who had high CGPA and more experience in language those who had high expectations for average salary.


```python
df['experience with python (months)'].value_counts()
# now to set the threshold for high experience - greater than or equal to 6 months.
high_cgpa_threshold = 8.0
high_experience_threshold = 6  # in months

# Filter individuals with high CGPA and more Python experience
high_cgpa_experience = df[(df['cgpa'] >= high_cgpa_threshold) & 
                          (df['experience with python (months)'] >= high_experience_threshold)]

# Plot Expected Salary vs CGPA for high CGPA and more experienced group
plt.figure(figsize=(10, 6))
sns.scatterplot(data=high_cgpa_experience, x='cgpa', y='expected salary (lac)', hue='experience with python (months)', palette='viridis')
plt.title('Expected Salary vs CGPA for High CGPA and More Experienced Individuals')
plt.xlabel('CGPA')
plt.ylabel('Expected Salary (in Lacs)')
plt.show()

# Plot Expected Salary vs Experience for high CGPA and more experienced group
plt.figure(figsize=(10, 6))
sns.scatterplot(data=high_cgpa_experience, x='experience with python (months)', y='expected salary (lac)', hue='cgpa', palette='viridis')
plt.title('Expected Salary vs Experience with Python for High CGPA Individuals')
plt.xlabel('Experience with Python (Months)')
plt.ylabel('Expected Salary (in Lacs)')
plt.show()
```

Solution:
While some individuals with high CGPA and Python experience have high salary expectations, there is no strong or consistent pattern in the data. The expected salary shows wide variability regardless of CGPA or Python experience. Therefore, it is not necessarily true that individuals with higher CGPA and more Python experience always have higher salary expectations.
